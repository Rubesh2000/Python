```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Měsíc': ['Leden', 'Únor', 'Březen', 'Duben'],
    'Hodnota': [2, 4, 10, 15]
}

df = pd.DataFrame(data)

#df.plot(kind='line', x='Měsíc', y='Hodnota')

# Spojitý graf 
plt.plot(data['Měsíc'], data['Hodnota'], label='Průměrné teploty v měsících')
plt.title('Průměrná teplota v měsících')
plt.xlabel('Měsíc')
plt.ylabel('Stupeň celsia')

# Vykreslím legendu
plt.legend()
# Vykreslím graf
plt.show()
```

```python
# Sloupcový graf
plt.bar(data['Měsíc'], data['Hodnota'], label='Průměrné teploty v měsících')
plt.title('Průměrná teplota v měsících')
plt.xlabel('Měsíc')
plt.ylabel('Stupeň celsia')

# Vykreslím legendu
plt.legend()
# Vykreslím mřížku
plt.grid()
# Vykreslím graf
plt.show()
```

```python
# Import z csv
df = pd.read_csv('Population in Poland - projections_1.csv', sep=';')

# Přetypování
df['Total'] = df['Total'].str.replace(' ', '').astype(int)
df['Men'] = df['Men'].str.replace(' ', '').astype(int)
df['Women'] = df['Women'].str.replace(' ', '').astype(int)
df.info()
```

```python
df.head(1)
```

```python
# Získat data pouze pro rok 2021
df_2021 = df.loc[ df['Year'] == 2021 ]

# Do grafu vytáhnout počet mužů v roce 2021
import matplotlib.pyplot as plt

# 1. graf - muži
plt.figure(figsize=(15,6))
plt.bar(df_2021['Age'], df_2021['Men'], color='green')
# Přidat název, pojmenování os a otočit labely na ose X o 45 stupňů
plt.title('Mužská populace v Polsku 2021 dle věkových skupin')
plt.xlabel('Věková skupiny')
plt.ylabel('Populace [mil]')
plt.xticks(rotation=45)
plt.grid()
# Exportuji do obrázku
plt.savefig(fname='muj_prvni_graf_muzi')
# Vykresluju graf
plt.show()

# 2. graf - ženy
plt.figure(figsize=(15,6))
plt.bar(df_2021['Age'], df_2021['Women'], color='red')
plt.title('Ženská populace v Polsku 2021 dle věkových skupin')
plt.xlabel('Věková skupiny')
plt.ylabel('Populace [mil]')
plt.xticks(rotation=45)
plt.grid()
# Exportuji do obrázku
plt.savefig(fname='muj_prvni_graf_zeny')
# Vykresluju graf
plt.show()
```

```python
df_2021.head()
```

```python
# Graf populace pro muže 2021
plt.figure(figsize=(15,8))

# Graf muži
plt.bar(df_2021['Age'], df_2021['Men'], label='Mužská populace 2021')
# Graf ženy
plt.bar(df_2021['Age'], df_2021['Women'], label='Ženská populace 2021', bottom=df_2021['Men'])
plt.title('Populace 2021 - Muži x ženy')
plt.xlabel('Věková skupina')
plt.ylabel('Populace [mil]')
plt.grid()
plt.legend()

plt.savefig('muziVSzeny_2021')
plt.show()
```

```python
# Získat data pouze pro rok 2021
df_2021 = df.loc[ df['Year'] == 2021 ]

# Do grafu vytáhnout počet mužů v roce 2021
import matplotlib.pyplot as plt

width = 0.25

# 1. graf - muži
plt.figure(figsize=(15,6))
plt.bar(df_2021['Age'].index, df_2021['Men'], color='green', width=0.25, label='Mužská populace')
# Přidat název, pojmenování os a otočit labely na ose X o 45 stupňů
plt.title('Mužská populace v Polsku 2021 dle věkových skupin')
plt.xlabel('Věková skupiny')
plt.ylabel('Populace [mil]')
plt.xticks(rotation=45)
plt.grid()

# 2. graf - ženy
plt.bar(df_2021['Age'].index+0.25, df_2021['Women'], color='red', width=0.25, label='Ženská populace')
plt.title('Populace v Polsku 2021 dle věkových skupin - ženy VS muži')
plt.xlabel('Věková skupiny')
plt.ylabel('Populace [mil]')
plt.xticks(ticks=df_2021['Age'].index, 
           labels=df_2021['Age'], 
           rotation=45)
plt.grid()
plt.legend()
# Exportuji do obrázku
plt.show()
```

```python
seskupena_data = df.groupby('Month')

df_humidity_month = seskupena_data['Average daily relative humidity'].mean().reset_index(name='monthly average humidity')
df_temperature_month = seskupena_data['Average daily temperature'].mean().reset_index(name='monthly average temperature')
df_wind_speed_month = seskupena_data['Average daily wind speed [m/s]'].mean().reset_index(name='monthly average wind speed')
df_cloud_cover_month = seskupena_data['Average daily cloud cover [oktas]'].mean().reset_index(name='monthly average cloud cover')
```

```python
seskup_data = df.groupby('Month')[['Average daily relative humidity', 
                                   'Average daily temperature', 
                                   'Average daily wind speed [m/s]', 
                                   'Average daily cloud cover [oktas]']].mean().reset_index()
seskup_data
```

```python
plt.figure(figsize=(20,10))
# Vlhkost
plt.subplot(2, 2, 1)
plt.plot(seskup_data['Month'], seskup_data['Average daily relative humidity'], label='Průměrná měsíční vlhkost')
plt.title('Průměrná vlhkost/měsíc')
plt.xlabel('Měsíc')
plt.ylabel('Vlhkost [%]')
plt.axhline(y=df['Average daily relative humidity'].mean(), 
            color='red',
            linestyle='--',
            label='Roční průměrná vlhkost')
plt.legend()

plt.xlabel('Měsíc')
plt.ylabel('Vlhkost [%]')
plt.axhline(y=df['Average daily relative humidity'].mean(),
color='red',
linestyle='--',
label='Roční průměrná vlhkost')


# Teplota
plt.subplot(2, 2, 2)
plt.plot(seskup_data['Month'], seskup_data['Average daily temperature'])
plt.title('Průměrná teplota / měsíc')
plt.xlabel('Měsíc')
plt.ylabel('Teplota [celsius]')
# Rychlost větru
plt.subplot(2, 2, 3)
plt.plot(seskup_data['Month'], seskup_data['Average daily wind speed [m/s]'])
plt.title('Průměrná rychlost větru / měsíc')
plt.xlabel('Měsíc')
plt.ylabel('Rychlost větru [m/s]')
# Pokrytí oblohy mraky
plt.subplot(2, 2, 4)
plt.plot(seskup_data['Month'], seskup_data['Average daily cloud cover [oktas]'])
plt.title('Průměrné pokrytí oblohy mraky/měsíc')
plt.xlabel('Měsíc')
plt.ylabel('Pokrytí oblohy mraky')

plt.show()
```
